```python
import pandas as pd
import glob
import os
```

##### List all the CSV file

```python
files = [f for f in glob.glob("*.csv") if f != "sales_all.csv"]
```

##### Combine all the CSV file

```python
# Only 1 header
df = pd.concat([pd.read_csv(f, header=0) for f in files], ignore_index=True)

# Remove rows where all columns are identical to the column headers
df = df[~(df == df.columns).all(axis=1)]

df
```

##### Check if there are blank rows

```python
blank_rows = df[df.isna().all(axis=1)]
blank_rows
```

##### Delete all blank rows

```python
df = df.dropna(how='all')
df
```

##### Rename column header

```python
df.columns = ['order_id', 'product', 'quantity_ordered', 'price_each', 'order_datetime', 'purchase_address']
df.head()
```

##### Check duplicates by order_id

```python
duplicates = df[df.duplicated(subset=['order_id'])]
duplicates
```

##### Delete duplicates by order_id

```python
df = df.drop_duplicates(subset=['order_id'], keep='first')
df
```

##### Reset Index

```python
df = df.reset_index(drop=True)
df.head()
```

##### Rename the 'index' column into 'No.'

```python
df.rename(columns={'index': 'No.'}, inplace=True)
df.head()
```

##### Separate date and time

```python
# Format date and time into the same format
df['order_datetime'] = pd.to_datetime(df['order_datetime'], format='%m/%d/%y %H:%M', errors='coerce')

# Separate date -> yyyy-mm-dd
df['order_date'] = df['order_datetime'].dt.date

# Separate month -> just the month
df['order_month'] = df['order_datetime'].dt.month

# Add quarter
df['order_quarter'] = pd.cut(
    df['order_month'], 
    bins=[0, 3, 6, 9, 12],
    labels=['Q1', 'Q2', 'Q3', 'Q4']
)

# Separate time
df['order_time'] = df['order_datetime'].dt.strftime('%H:%M')

# Delete the 'order_datetime' column
df.drop('order_datetime', axis=1, inplace=True)

# Check the table
df
```

##### Separate address into street, city, state, zip

```python
# Separate street, city, state_zip
df[['street', 'city', 'state_zip']] = df['purchase_address'].str.split(',', expand=True)

# Separate state and zip from 'state_zip' column
df[['state','zip']] = df['state_zip'].str.strip().str.split(' ', n=1, expand=True)

# Delete the 'purchase_address' and 'state_zip' column
df.drop(['purchase_address','state_zip'], axis=1, inplace=True)

# Check the table
df
```

##### Save the final CSV:
